# import libraries

```python
import pandas as pd
```

# import data

```python
df = pd.read_csv("./salaries.csv")
df.head()
df.shape
df.columns
```

# Perliminary Data cleaning

```python
clean_df = df.dropna()
clean_df.tail()
```

# Finding the Major with Highest Starting Salary

```python
max_id = clean_df['Starting Median Salary'].idxmax()
max_major = clean_df['Undergraduate Major'].loc[max_id]
print(max_major)
```

# Finding the Major with Highest Median Salary

```python
max_id = clean_df['Mid-Career Median Salary'].idxmax()
max_major = clean_df.loc[max_id]
print(max_major)
```


# Finding the Major with Lowest Starting Salary

```python
max_id = clean_df['Starting Median Salary'].idxmin()
max_major = clean_df.loc[max_id]
print(max_major)
```

# Finding the Major with Lowest Median Salary

```python
max_id = clean_df['Mid-Career Median Salary'].idxmin()
max_major = clean_df.loc[max_id]
print(max_major)
```

# Calculation Spread

```python
spread = abs(clean_df['Mid-Career 10th Percentile Salary'] - clean_df['Mid-Career 90th Percentile Salary'])
clean_df.insert(1 , 'Spread' , spread)
```

```python
clean_df.head()
```

# Low risk majors

```python
low_risk = clean_df.sort_values('Spread').head(5)
```

```python
print(low_risk)
```

# Jobs with highest potential

```python
high_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False).head(5)
```

```python
print(high_potential)
```

# Highest Risk Jobs

```python
high_risk = clean_df.sort_values('Spread', ascending=False).head(5)
```

```python
print(high_risk)
```

# Grouping by most Group

```python
most_group = clean_df.groupby('Group').count()
```

```python
print(most_group)
```

# mean salary by group 

```python
groups = list(clean_df['Group'].unique())
```

```python
print(groups)
```

```python
# getting mean salary per group
mean_salaries = {}
for group in groups:
    mean_salaries[group] = float(clean_df[clean_df['Group']==group]['Mid-Career Median Salary'].mean())
```

```python
mean_df = clean_df.groupby('Group').mean(numeric_only=True)
print(mean_df)
```

```python
print(mean_salaries)
```
